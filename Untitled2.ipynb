```python

detected = []

for r in results:
    for box in r.boxes:

        cls = int(box.cls)

        name = model.names[cls]

        detected.append(name)

        print(name)
```

```python
from collections import Counter

counter = Counter(detected)

print("Electronic Components Detected\n")

for component, count in counter.items():
    print(f"{component}: {count}")
```
